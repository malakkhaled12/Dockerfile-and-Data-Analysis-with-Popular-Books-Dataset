```python
import pandas as pd

# Specify the filename of the uploaded dataset (e.g., 'books.csv')
uploaded_file = 'books.csv'  # Update with the actual filename

# Load the dataset using the filename of the uploaded file
df = pd.read_csv(uploaded_file)

# Display the first few rows of the dataset to verify
df.head()
```

```python
df.info()

# Summary statistics of numerical columns
df.describe()
```

```python
harry_potter_books = df[df['title'].str.contains('Harry Potter', case=False, na=False)]

# Check the filtered DataFrame to ensure it contains Harry Potter books
print(harry_potter_books)
```

```python
average_rating = harry_potter_books['average_rating'].mean()
print(f"Average rating of Harry Potter books: {average_rating:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Create a bar plot of book ratings
sns.barplot(x='title', y='average_rating', data=harry_potter_books)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title('Rating of Harry Potter Books')
plt.xlabel('Book Title')
plt.ylabel('Rating')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Assuming 'harry_potter_books' is your DataFrame containing the Harry Potter book series data

# Group by book title and calculate the total ratings count for each book
total_ratings_per_book = harry_potter_books.groupby('title')['ratings_count'].sum().reset_index()

# Sort the books based on total ratings count in descending order
sorted_books_by_ratings = total_ratings_per_book.sort_values(by='ratings_count', ascending=False)

# Display the top-rated Harry Potter books based on total ratings count
print("Top-rated Harry Potter Books by Total Ratings Count:")
print(sorted_books_by_ratings.head(10))  # Display the top 10 books (adjust as needed)

# Plotting the top-rated books by total ratings count
plt.figure(figsize=(12, 8))
sns.barplot(data=sorted_books_by_ratings.head(10), x='ratings_count', y='title', hue='ratings_count', dodge=False)
plt.title('Top-rated Harry Potter Books by Total Ratings Count')
plt.xlabel('Total Ratings Count')
plt.ylabel('Book Title')
plt.legend().remove()  # Remove the legend (optional)
plt.show()
```
